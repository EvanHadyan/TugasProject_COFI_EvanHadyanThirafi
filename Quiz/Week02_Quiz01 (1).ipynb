### **Soal 3 List Comprehension & Pandas** 

Pada Soal ini kalian **wajib** menggunakan feature list comprehension dan juga apply.<br>

```python
import pandas as pd
df=pd.read_csv('Salaries.csv')
df
```

PT.A baru saja kena tegur dari negara karena ketauan tidak membayar uang lemburan. Anda sebagai seorang data Scientist diminta PT.A untuk mengolah data yang ada.

**Step pertama**

**Wajib Menggunakan .Apply()** <br>
PT.A melihat beberapa data tidak di isi oleh sebagian karyawan. kalian di minta untuk untuk merubah nilai 'Not provided' pada column `JobTitle` dan `EmployeeName` menjadi 'Error'.

```python
def word(a):
    if 'Not provided' in a:
       return 'Error'
    else:
       return a


df['JobTitle'] = df['JobTitle'].apply(word)
df['EmployeeName'] = df['EmployeeName'].apply(word)
```

### **Soal 4 One Hot Encoding**

Buatlah **sebuah fungsi** bernama `one_hot`, yang menerima input sebuah numpy array.

Ex:<br>
input array [2,3,1,2] maka hasil dari fungsi kalian:

array [[0,1,0], [0,0,1], [1,0,0], [0,1,0]]

Constrain: <br>

- Integers dalam input selalu di awali 1. 
- Integers dalam input hanya positif dan bukan 0.
- Integers dalam input tidak akan skip angka.

```python
import numpy as np
input_array = np.array([3,2,1,2,3,3,3])
#your code goes here
b = np.zeros((input_array.size, input_array.max() + 1))
b[np.arange(input_array.size), input_array] = 1
```
